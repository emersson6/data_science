```python
### Load relevant packages
import pandas                  as pd
from   scipy import stats
import numpy                   as np
import matplotlib.pyplot       as plt
import seaborn                 as sns
import statsmodels.formula.api as sm
import chart_studio.plotly     as py
import seaborn.objects as so

# https://community.plot.ly/t/solved-update-to-plotly-4-0-0-broke-application/26526/2
import os

#%matplotlib inline
#plt.style.use('ggplot')
from bokeh.resources import INLINE
import bokeh.io
from bokeh import *
import pingouin
bokeh.io.output_notebook(INLINE)


url= 'https://raw.githubusercontent.com/emersson6/data_science/main/Desafio_3/datos_de_venta_3.csv'
df = pd.read_csv(url, index_col=0, sep=';')

df['product_quantity'] = df['product_quantity'].astype(int)
df['total_price_tax_incl'] = df['total_price_tax_incl'].astype(int)
df['Fecha'] = pd.to_datetime(df['Fecha'])
df['Year'] = df['Fecha'].dt.year
df['Month'] =df['Fecha'].dt.month

df.head()
```

```python
df_calidad=df[['Year','Marca','Calidad_Cepa','product_quantity','Month']].groupby(by=['Year','Month','Marca','Calidad_Cepa']).sum().reset_index()
df_calidadx=df_calidad[['Month','Year','product_quantity']].groupby(by=['Month','Year']).sum().reset_index()
```

# Distribución de ventas por mes

Este gráfico, nos muestras en que meses se genera la mayor cantidad de ventas y estas se focalizan principalmente agosto y diciembre.

```python
sns.set_theme()
fig, ax = plt.subplots(figsize=(15, 7))
sns.lineplot(x="Month", y="product_quantity", data=df_calidadx, hue="Year", style="Year",markers=True, dashes=False) 
plt.title("Distribucion por año");
```

# Exploracion de las ciudades que generan compras

```python
print(df['Ciudad'].unique())
```

Se analizan las distintas ciudades desde donde se realizan compras. Se observa una gran cantidad de diferentes puntos de compra

```python
Ciudades=df[['Ciudad','Total Pagado']].groupby(by=['Ciudad']).count().reset_index()
Ciudades.rename(columns={'Total Pagado':'Cantidad de compras'},inplace=True)
porcantidad=Ciudades.sort_values('Cantidad de compras',ascending=False)
porcantidad.head()
```

# Total de ciudades

```python
Ciudades['Cantidad de compras'].sum()
```

```python
porcantidad['porcentajedecompras']=porcantidad['Cantidad de compras']*100/37292
porcantidad.head()
```

```python
sns.barplot(x="Ciudad", y="porcentajedecompras",data=porcantidad, ) 
sns.set( rc = {'figure.figsize' :( 200, 150 ),'axes.labelsize' : 100 })
plt.title("Cantida de compras por ciudad");
```

Del analisis de las compras, se verifica que el 20% de las compras se realiza desde la ciudad de santiago. Adicionalmente que aproximadamnete el 60% de las compras totales se realiza unicamente desde 5 localicadaes de las aproximadamente 170 localidades totales que existen en la base de datos
